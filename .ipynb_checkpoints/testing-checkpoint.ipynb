```python
import json

# Load the JSON data
file_path = "/Users/vsai23/Workspace/MBQC2/QCBacktesting/qms/BSMR.json"
with open(file_path, 'r') as f:
    data = json.load(f)

# Initialize sets for indicators and symbols
indicators = set()
symbols = set()

def parse_condition(condition):
    if "lh_indicator" in condition:
        lh_indicator = condition["lh_indicator"]["type"]
        lh_period = condition["lh_indicator"].get("window", None)
        indicators.add((lh_indicator, lh_period))
    if "rh_indicator" in condition:
        rh_indicator = condition["rh_indicator"]["type"]
        rh_period = condition["rh_indicator"].get("window", None)
        indicators.add((rh_indicator, rh_period))
    if "lh_ticker_symbol" in condition and condition["lh_ticker_symbol"]:
        symbols.add(condition["lh_ticker_symbol"])
    if "rh_ticker_symbol" in condition and condition["rh_ticker_symbol"]:
        symbols.add(condition["rh_ticker_symbol"])

def parse_incantation(incantation):
    if incantation:
        if "condition" in incantation:
            parse_condition(incantation["condition"])
        if "else_incantation" in incantation and incantation["else_incantation"]:
            parse_incantation(incantation["else_incantation"])
        if "then_incantation" in incantation and incantation["then_incantation"]:
            if incantation["then_incantation"]["incantation_type"] == "Ticker":
                symbols.add(incantation["then_incantation"]["symbol"])
            else:
                parse_incantation(incantation["then_incantation"])

# Parse the top-level incantation
parse_incantation(data["incantation"])

# Convert sets to lists
indicator_list = list(indicators)
symbol_list = list(symbols)

# Print the results
print("Indicators:", indicator_list)
print("Symbols:", symbol_list)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data as list of dictionaries
data = [
    {"id": "1575129947-1-1", "algorithmId": "1575129947", "orderId": 1, "orderEventId": 1, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1665690960.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "buy", "isAssignment": False, "quantity": 5497.0},
    {"id": "1575129947-1-2", "algorithmId": "1575129947", "orderId": 1, "orderEventId": 2, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1665690960.0, "status": "filled", "orderFeeAmount": 27.485, "orderFeeCurrency": "USD", "fillPrice": 18.14, "fillPriceCurrency": "USD", "fillQuantity": 5497.0, "direction": "buy", "message": "Warning: No quote information available at 10/13/2022 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": 5497.0},
    {"id": "1575129947-2-1", "algorithmId": "1575129947", "orderId": 2, "orderEventId": 1, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1666036560.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "sell", "isAssignment": False, "quantity": -5497.0},
    {"id": "1575129947-2-2", "algorithmId": "1575129947", "orderId": 2, "orderEventId": 2, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1666036560.0, "status": "filled", "orderFeeAmount": 27.485, "orderFeeCurrency": "USD", "fillPrice": 19.31, "fillPriceCurrency": "USD", "fillQuantity": -5497.0, "direction": "sell", "message": "Warning: No quote information available at 10/15/2022 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": -5497.0},
    {"id": "1575129947-3-1", "algorithmId": "1575129947", "orderId": 3, "orderEventId": 1, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1666036560.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "buy", "isAssignment": False, "quantity": 167.0},
    {"id": "1575129947-3-2", "algorithmId": "1575129947", "orderId": 3, "orderEventId": 2, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1666036560.0, "status": "filled", "orderFeeAmount": 1.0, "orderFeeCurrency": "USD", "fillPrice": 634.5, "fillPriceCurrency": "USD", "fillQuantity": 167.0, "direction": "buy", "message": "Warning: No quote information available at 10/15/2022 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": 167.0},
    {"id": "1575129947-4-1", "algorithmId": "1575129947", "orderId": 4, "orderEventId": 1, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1666122960.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "sell", "isAssignment": False, "quantity": -167.0},
    {"id": "1575129947-4-2", "algorithmId": "1575129947", "orderId": 4, "orderEventId": 2, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1666122960.0, "status": "filled", "orderFeeAmount": 1.0, "orderFeeCurrency": "USD", "fillPrice": 634.5, "fillPriceCurrency": "USD", "fillQuantity": -167.0, "direction": "sell", "message": "Warning: No quote information available at 10/15/2022 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": -167.0},
    {"id": "1575129947-5-1", "algorithmId": "1575129947", "orderId": 5, "orderEventId": 1, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1666122960.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "buy", "isAssignment": False, "quantity": 5493.0},
    {"id": "1575129947-5-2", "algorithmId": "1575129947", "orderId": 5, "orderEventId": 2, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1666122960.0, "status": "filled", "orderFeeAmount": 27.465, "orderFeeCurrency": "USD", "fillPrice": 19.31, "fillPriceCurrency": "USD", "fillQuantity": 5493.0, "direction": "buy", "message": "Warning: No quote information available at 10/15/2022 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": 5493.0},
    {"id": "1575129947-6-1", "algorithmId": "1575129947", "orderId": 6, "orderEventId": 1, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1687290960.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "sell", "isAssignment": False, "quantity": -5493.0},
    {"id": "1575129947-6-2", "algorithmId": "1575129947", "orderId": 6, "orderEventId": 2, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1687290960.0, "status": "filled", "orderFeeAmount": 27.465, "orderFeeCurrency": "USD", "fillPrice": 41.6, "fillPriceCurrency": "USD", "fillQuantity": -5493.0, "direction": "sell", "message": "Warning: No quote information available at 06/17/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": -5493.0},
    {"id": "1575129947-7-1", "algorithmId": "1575129947", "orderId": 7, "orderEventId": 1, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1687290960.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "buy", "isAssignment": False, "quantity": 2028.0},
    {"id": "1575129947-7-2", "algorithmId": "1575129947", "orderId": 7, "orderEventId": 2, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1687290960.0, "status": "filled", "orderFeeAmount": 10.14, "orderFeeCurrency": "USD", "fillPrice": 112.5, "fillPriceCurrency": "USD", "fillQuantity": 2028.0, "direction": "buy", "message": "Warning: No quote information available at 06/17/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": 2028.0},
    {"id": "1575129947-8-1", "algorithmId": "1575129947", "orderId": 8, "orderEventId": 1, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1687377360.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "sell", "isAssignment": False, "quantity": -2028.0},
    {"id": "1575129947-8-2", "algorithmId": "1575129947", "orderId": 8, "orderEventId": 2, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1687377360.0, "status": "filled", "orderFeeAmount": 10.14, "orderFeeCurrency": "USD", "fillPrice": 112.5, "fillPriceCurrency": "USD", "fillQuantity": -2028.0, "direction": "sell", "message": "Warning: No quote information available at 06/17/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": -2028.0},
    {"id": "1575129947-9-1", "algorithmId": "1575129947", "orderId": 9, "orderEventId": 1, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1687377360.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "buy", "isAssignment": False, "quantity": 5484.0},
    {"id": "1575129947-9-2", "algorithmId": "1575129947", "orderId": 9, "orderEventId": 2, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1687377360.0, "status": "filled", "orderFeeAmount": 27.42, "orderFeeCurrency": "USD", "fillPrice": 41.6, "fillPriceCurrency": "USD", "fillQuantity": 5484.0, "direction": "buy", "message": "Warning: No quote information available at 06/17/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": 5484.0},
    {"id": "1575129947-10-1", "algorithmId": "1575129947", "orderId": 10, "orderEventId": 1, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1689882960.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "sell", "isAssignment": False, "quantity": -5484.0},
    {"id": "1575129947-10-2", "algorithmId": "1575129947", "orderId": 10, "orderEventId": 2, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1689882960.0, "status": "filled", "orderFeeAmount": 27.42, "orderFeeCurrency": "USD", "fillPrice": 46.32, "fillPriceCurrency": "USD", "fillQuantity": -5484.0, "direction": "sell", "message": "Warning: No quote information available at 07/20/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": -5484.0},
    {"id": "1575129947-11-1", "algorithmId": "1575129947", "orderId": 11, "orderEventId": 1, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1689882960.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "buy", "isAssignment": False, "quantity": 2999.0},
    {"id": "1575129947-11-2", "algorithmId": "1575129947", "orderId": 11, "orderEventId": 2, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1689882960.0, "status": "filled", "orderFeeAmount": 14.995, "orderFeeCurrency": "USD", "fillPrice": 84.65, "fillPriceCurrency": "USD", "fillQuantity": 2999.0, "direction": "buy", "message": "Warning: No quote information available at 07/20/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": 2999.0},
    {"id": "1575129947-12-1", "algorithmId": "1575129947", "orderId": 12, "orderEventId": 1, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1690228560.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "sell", "isAssignment": False, "quantity": -2999.0},
    {"id": "1575129947-12-2", "algorithmId": "1575129947", "orderId": 12, "orderEventId": 2, "symbol": "UVXY 2T", "symbolValue": "UVXY", "symbolPermtick": "UVXY", "time": 1690228560.0, "status": "filled", "orderFeeAmount": 14.995, "orderFeeCurrency": "USD", "fillPrice": 87.15, "fillPriceCurrency": "USD", "fillQuantity": -2999.0, "direction": "sell", "message": "Warning: No quote information available at 07/22/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": -2999.0},
    {"id": "1575129947-13-1", "algorithmId": "1575129947", "orderId": 13, "orderEventId": 1, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1690228560.0, "status": "submitted", "fillPrice": 0.0, "fillPriceCurrency": "USD", "fillQuantity": 0.0, "direction": "buy", "isAssignment": False, "quantity": 6068.0},
    {"id": "1575129947-13-2", "algorithmId": "1575129947", "orderId": 13, "orderEventId": 2, "symbol": "TQQQ 2T", "symbolValue": "TQQQ", "symbolPermtick": "TQQQ", "time": 1690228560.0, "status": "filled", "orderFeeAmount": 30.34, "orderFeeCurrency": "USD", "fillPrice": 43.07, "fillPriceCurrency": "USD", "fillQuantity": 6068.0, "direction": "buy", "message": "Warning: No quote information available at 07/22/2023 00:00:00 America/New_York, order filled using TradeBar data", "isAssignment": False, "quantity": 6068.0}
]

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Convert time to datetime
df['time'] = pd.to_datetime(df['time'], unit='s')

# Calculate P/L for each trade
df['trade_value'] = df['fillPrice'] * df['fillQuantity']
df['direction_multiplier'] = df['direction'].apply(lambda x: 1 if x == 'buy' else -1)
df['pl'] = df['trade_value'] * df['direction_multiplier']

# Calculate cumulative P/L
df['cum_pl'] = df['pl'].cumsum()

# Calculate equity
initial_equity = 100000  # Example initial equity
df['equity'] = initial_equity + df['cum_pl']

# Calculate drawdown
df['peak_equity'] = df['equity'].cummax()
df['drawdown'] = df['equity'] - df['peak_equity']
df['drawdown_pct'] = df['drawdown'] / df['peak_equity'] * 100

# Plotting the charts

# Equity chart
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['equity'], label='Equity')
plt.xlabel('Time')
plt.ylabel('Equity ($)')
plt.title('Equity Curve')
plt.legend()
plt.grid(True)
plt.show()

# Drawdown chart
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['drawdown'], label='Drawdown ($)')
plt.xlabel('Time')
plt.ylabel('Drawdown ($)')
plt.title('Drawdown ($)')
plt.legend()
plt.grid(True)
plt.show()

# Drawdown percentage chart
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['drawdown_pct'], label='Drawdown (%)')
plt.xlabel('Time')
plt.ylabel('Drawdown (%)')
plt.title('Drawdown (%)')
plt.legend()
plt.grid(True)
plt.show()

# P/L percentage chart
df['pl_pct'] = df['pl'] / initial_equity * 100
df['cum_pl_pct'] = df['pl_pct'].cumsum()

plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['cum_pl_pct'], label='P/L (%)')
plt.xlabel('Time')
plt.ylabel('P/L (%)')
plt.title('P/L Percentage')
plt.legend()
plt.grid(True)
plt.show()

# Bar chart of trades
trade_counts = df.groupby('symbolValue').size()

plt.figure(figsize=(12, 6))
trade_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Symbol')
plt.ylabel('Number of Trades')
plt.title('Number of Trades per Symbol')
plt.grid(True)
plt.show()
```
